```julia
using DelimitedFiles

# Read the MEI index
mei = readdlm("../data/MEI_1951_2016.txt")

# Read the PNA index
pna = readdlm("../data/PNA_1951_2016.txt")
```

```julia
# Read through data and create time and array with both indices but only use the January values
time = mei[:, 1]
val = Array{Float64, 2}(undef, length(time), 2)
val[:, 1] = mei[:, 2]
val[:, 2] = pna[:, 2]
```

```julia
# Create std anom of indices
using Statistics
val[:, 1] = (val[:, 1] .- mean(val[:, 1])) / std(val[:, 1], corrected=false)
val[:, 2] = (val[:, 2] .- mean(val[:, 2])) / std(val[:, 2], corrected=false);
```

```julia
using Plots
plotly()
plot(time, val[:, 1], color="red", xlabel="Time", ylabel="MEI-r PNA-b", label="MEI",
     title="MEI and PNA Indices: Alex Weech 2/9/19", xlims=(1950, 2021), ylims=(-3, 3))
plot!(time, val[:, 2], color="blue", label="PNA")
```

```julia
# Compute anova table
r = cor(val)
r2 = r[1, 2]^2
n = length(val[:, 1])
SST = n
SSR = n * r2
SSE = n * (1-r2)
MST = n / (n-1)
MSR = n*r2
MSE = n * (1-r2)/(n-2)
F = (n-2)*r2/(1-r2);
```
